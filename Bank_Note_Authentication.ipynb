# Step-1) Business problem Understanding

## Bank Note Authentication
Data were extracted from images that were taken from genuine and forged banknote-like specimens. For digitization, an industrial camera usually used for print inspection was used. The final images have 400x 400 pixels. Due to the object lens and distance to the investigated object gray-scale pictures with a resolution of about 660 dpi were gained. Wavelet Transform tool were used to extract features from images.

### 3.3.4) Train-Test Split

```python
### Train Test Split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

# Step-4) Modeling

# 1) Logistic Regression

```python
# Modeling
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
log_model.fit(x_train,y_train)

# Prediction
ypred_train = log_model.predict(x_train)
ypred_test = log_model.predict(x_test)

# Evaluation
# 1.train-test accuracy
from sklearn.metrics import accuracy_score
print(accuracy_score(y_train,ypred_train))   #Train accuracy
print(accuracy_score(y_test,ypred_test))   #Test accuracy

#2.cross val score
from sklearn.model_selection import cross_val_score
scores = cross_val_score(log_model,x,y,cv=5)

print(scores)
print("cross val score is:",scores.mean())
```

# Creating ML PIPELINE

```python
from sklearn.pipeline import Pipeline
```

```python
LogisticRegressionPipeline = Pipeline([("logistic_classifier", LogisticRegression())])

KNeighborsClassifierPipeline = Pipeline([("knn_classifier", KNeighborsClassifier())])

SVCPipeline = Pipeline([("svm_classifier", SVC())])

DecisionTreeClassifierPipeline = Pipeline([("decisiontree_classifier", DecisionTreeClassifier())])

RandomForestClassifierPipeline = Pipeline([("randomforest_classifier", RandomForestClassifier())])

AdaBoostClassifierPipeline = Pipeline([("adaboost_classifier", AdaBoostClassifier())])

GradientBoostingClassifierPipeline = Pipeline([("gradientboost_classifier", GradientBoostingClassifier(n_estimators=30, learning_rate=1))])

XGBClassifierPipeline = Pipeline([("xg_boost_classifier", XGBClassifier())])
```

# Modeling & Evaluation

```python
# defining a pipeline in the list
mypipeline = [LogisticRegressionPipeline, KNeighborsClassifierPipeline, SVCPipeline, DecisionTreeClassifierPipeline, RandomForestClassifierPipeline, AdaBoostClassifierPipeline, GradientBoostingClassifierPipeline, XGBClassifierPipeline]
```

```python
# fit the pipelines
for mypipe in mypipeline:
    mypipe.fit(x_train, y_train)
```

```python
# defining variables for choosing best models
accuracy = 0.0
classifier = 0
pipeline = ""
```

```python
# creating dictionary of pipelines and training models
pipelinedict = {0:"Logistic Regression", 1:"K-NN Classifier", 2:"SVC", 3:"Decision Tree", 4:"Random Forest Classifier", 5:"Ada Boost", 6:"Gradient Boost", 7:"XG Boost Classifier"}
```

```python
# getting test accuracy for all classifier
for i, model in enumerate(mypipeline):
    if model.score(x_test, y_test)>accuracy:
        accuracy = model.score(x_test, y_test)
        pipeline = model
        classifier = i
print("Classifier with best accuracy :{}".format(pipelinedict[classifier]))
```
